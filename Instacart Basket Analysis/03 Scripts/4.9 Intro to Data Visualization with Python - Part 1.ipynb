# Completing the fundamental data quality and consistency checks

```python
#checking for missing values
df_cust.isnull().sum()
```

#### Note: blank values in the "First Name" is not an issue. I will not perform any action on these missing values

```python
# checking for duplicates in the customers data set
df_duplicates = df_cust[df_cust.duplicated()]
```

```python
df_duplicates
```

#### No results found. Thus no duplicates 

```python
# checking data types
data_types = df_cust.dtypes
print(data_types)
```

```python
# checking for mixed data types
for col in df_cust.columns.tolist():
  weird = (df_cust[[col]].applymap(type) != df_cust[[col]].iloc[0].apply(type)).any(axis = 1)
  if len (df_cust[weird]) > 0:
    print (col)
```

#### Based on the above, we need to change the data type to string in both 'first_name' and 'user_id' columns

```python
#changing the data type to string in user_id in df_cust
df_cust['user_id'] = df_cust['user_id'].astype('str')
```

```python
#changing the data type to string in first_name in df_cust
df_cust['first_name'] = df_cust['first_name'].astype('str')
```

# Combine customer data with the rest of prepared Instacart data

```python
# preparing for the merge: checking all columns of ords_prods_merge
print(ords_prods_merge.columns.tolist())
```

```python
# deleting the '_merge' column
ords_prods_merge = ords_prods_merge.drop(columns = ['_merge'])
```

```python
# checking data type of the key cloumn, 'user_id'
data_types = ords_prods_merge.dtypes
print(data_types)
```

```python
# changing the data type of "user_id" to string. This way it matches the 'user_id' data type in df_cust
ords_prods_merge['user_id'] = ords_prods_merge['user_id'].astype(str)
```

```python
# merging the df_cust to ords_prods_merge
df_cust_ords_prods_merge = df_cust.merge(ords_prods_merge, on = 'user_id', indicator = True)
```
