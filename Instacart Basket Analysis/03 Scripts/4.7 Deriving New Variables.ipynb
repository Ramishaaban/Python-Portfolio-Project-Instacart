## if statement with user-defined functions

```python
# create a price label with if-statements
def price_label(row):

  if row['prices'] <= 5:
    return 'Low-range product'
  elif (row['prices'] > 5) and (row['prices'] <= 15):
    return 'Mid-range product'
  elif row['prices'] > 15:
    return 'High range'
  else: return 'Not enough data'
```

```python
# apply the function
df['price_range'] = df.apply(price_label, axis=1)
```

```python
df['price_range'].value_counts(dropna = False)
```

## If-statement with for-loop

```python
# printing the frequency of the “orders_day_of_the_week” column
df_ords_prods_merged['orders_day_of_week'].value_counts(dropna = False)
```

```python
# create for-loop
result = []

for value in df_ords_prods_merged["orders_day_of_week"]:
  if value == 0:
    result.append("Busiest day")
  elif value == 4:
    result.append("Least busy")
  else:
    result.append("Regularly busy")
```

```python
result
```

```python
# create a new column within df_ords_prods_merged and set it equal to result
df_ords_prods_merged['busiest_day'] = result
```

```python
df_ords_prods_merged['busiest_day'].value_counts(dropna = False)
```

# Step 2 of the task: "Busiest days" and "slowest days"

```python
# Create a for-loop
plural_result = []

for value in df_ords_prods_merged["orders_day_of_week"]:
  if value in [0, 1]:
    plural_result.append("Busiest days")
  elif value in [3, 4]:
    plural_result.append("Slowest days")
  else:
    plural_result.append("Regularly busy")
```

```python
plural_result
```

```python
# create a new column within df_ords_prods_merged and set it equal to plural_result
df_ords_prods_merged['busiest_days'] = plural_result
```

# Step 4 of the task: finding busiest_period_of_day

```python
# printing the frequency of the “order_hour_of_day” column
df_ords_prods_merged['order_hour_of_day'].value_counts(dropna = False)
```

### the senior technical officer wants periods of time labeled “Most orders,” “Average orders,” and “Fewest orders.”. I suggest the three labels split as: Most orders: 10, 11, 14, 15, 13, 12, 16. Average orders: 9, 17, 8, 18, 19. Fewest orders:20, 7, 21, 22, 23, 6, 0, 1, 5, 2, 4, 3.

```python
# Create a for-loop
hours_result = []

for value in df_ords_prods_merged["order_hour_of_day"]:
  if value in [10, 11, 14, 15, 13, 12, 16]:
    hours_result.append("Most orders")
  elif value in [9, 17, 8, 18, 19]:
    hours_result.append("Average orders")
  else:
    hours_result.append("Fewest orders")
```

```python
# create a new column within df_ords_prods_merged and set it equal to busiest_period_of_day
df_ords_prods_merged['busiest_period_of_day'] = hours_result
```
