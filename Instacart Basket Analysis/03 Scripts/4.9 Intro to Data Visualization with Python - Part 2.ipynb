# Create a histogram of the “order_hour_of_day” column

```python
# Create a histogram for 'order_hour_of_day' column
hist_3 = df_cust_ords_prods_merge['order_hour_of_day'].plot.hist(bins = 24)
```

```python
# Saving the histogram to the Viz folder
hist_3.figure.savefig(os.path.join(path, '04 Analysis','Visualizations', 'hist3_order_hour_of_day.png'))
```

###  Description of the histogram
The histogram shows the distribution of the number of orders placed at different hours of the day.
The number of orders placed is relatively low in the early morning hours, starts to increase in the late morning hours, peaks in the hours between 9:00 to 16:00, and then starts to decrease in the evening hours. The lowest hours are between midnight and 6:00
